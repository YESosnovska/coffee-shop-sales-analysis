```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Coffee Shop Sales.csv")

data.head()
```

```python
data.info()
```

```python
data["transaction_date"] = pd.to_datetime(data["transaction_date"], format="%m/%d/%Y")

data.info()
```

```python
data["year"] = data["transaction_date"].dt.year

data["month"] = data["transaction_date"].dt.month_name()

data["day"] = data["transaction_date"].dt.day_name()

data.head()
```

```python
counts = data["day"].value_counts()
days = data["day"].unique()

plt.bar(days, counts)
plt.title("Sales by day")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
data["transaction_time"] = pd.to_datetime(data["transaction_time"], format='%H:%M:%S').dt.time

data.head()
```

```python
data["hour"] = data["transaction_time"].apply(lambda x : x.hour)

data.head()
```

```python
hour_counts = data["hour"].value_counts()
hours = data["hour"].unique()

fig, ax = plt.subplots()

ax.stem(hours, hour_counts)
ax.set_title("Sales by hour")
ax.set_xlabel("Hour")
ax.set_ylabel("Count")
plt.show()
```

```python
products = data["product_category"].unique()
product_count = data["product_category"].value_counts()

plt.bar(products, product_count)
plt.title("Most popular product categories")
plt.xlabel("Product category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
data["profit"] = data["transaction_qty"] * data["unit_price"]

data.head()
```

```python
profit_by_product = data.groupby("product_category")["profit"].sum().sort_values(ascending=False)

plt.bar(products, profit_by_product)
plt.title("Profit by category")
plt.xlabel("Product category")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
coffee_products = data[data["product_category"] == "Coffee"]
coffee_types = coffee_products["product_detail"].unique()
coffee_types_count = coffee_products["product_detail"].value_counts()

tea_products = data[data["product_category"] == "Tea"]
tea_types = tea_products["product_detail"].unique()
tea_types_count = tea_products["product_detail"].value_counts()

fig, (ax1, ax2) = plt.subplots(2, figsize=(10, 6))
fig.subplots_adjust(hspace=2.5)
ax1.bar(coffee_types, coffee_types_count)
ax2.bar(tea_types, tea_types_count)
ax1.set_title("Popular coffee products")
ax2.set_title("Popular tea products")
ax2.set_xlabel("Types")
ax1.set_ylabel("Count")
ax1.set_xticks(range(len(coffee_types)))
ax1.set_xticklabels(coffee_types, rotation=90)
ax2.set_xticks(range(len(tea_types)))
ax2.set_xticklabels(tea_types, rotation=90)
plt.show()
```

```python
store_location = data["store_location"].unique()
store_location_count = data["store_location"].value_counts()

fig, ax = plt.subplots()
ax.pie(store_location_count, labels=store_location, autopct='%1.1f%%', startangle=90)
plt.title("Store location popularity")
plt.show()
```

```python
profit_by_store = data.groupby("store_location")["profit"].sum().sort_values(ascending=False)

fig, ax = plt.subplots()
ax.pie(profit_by_store, labels=store_location, autopct='%1.1f%%', startangle=90)
plt.title("Profit by store location")
plt.show()
```
