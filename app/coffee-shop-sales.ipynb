```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Coffee Shop Sales.csv")

data.head()
```

```python
data.info()
```

```python
data["transaction_date"] = pd.to_datetime(data["transaction_date"], format="%m/%d/%Y")

data.info()
```

```python
data["year"] = data["transaction_date"].dt.year

data["month"] = data["transaction_date"].dt.month_name()

data["day"] = data["transaction_date"].dt.day_name()

data.head()
```

```python
counts = data["day"].value_counts()
days = data["day"].unique()

plt.bar(days, counts)
plt.title("Sales by day")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
data["transaction_time"] = pd.to_datetime(data["transaction_time"], format='%H:%M:%S').dt.time

data.head()
```

```python
data["hour"] = data["transaction_time"].apply(lambda x : x.hour)

data.head()
```

```python
hour_counts = data["hour"].value_counts()
hours = data["hour"].unique()

fig, ax = plt.subplots()

ax.stem(hours, hour_counts)
ax.set_title("Sales by hour")
ax.set_xlabel("Hour")
ax.set_ylabel("Count")
plt.show()
```
